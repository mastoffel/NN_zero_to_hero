Get all Sherlock Holmes stories and concatenate them into a single text.

```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
base_url = "https://sherlock-holm.es/"
stories_url = "/ascii/"
response = requests.get(base_url + stories_url)
```

```python
# BeautifulSoup object with html parser
soup = BeautifulSoup(response.text, "html.parser")
```

```python
links = soup.find_all('a')
links
```

```python

# filter for txt files
story_links = [base_url + link['href'] for link in links if link['href'].endswith('.txt')]
```

```python
story_links
```

```python
story_links[0]
```

```python
with open('sherlock.txt', 'w') as outfile:
    for link in story_links:
        response = requests.get(link)
        outfile.write(response.text)
        outfile.write('\n')
        
```
